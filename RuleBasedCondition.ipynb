```python
def result_condition(line):
    if line['Age']<24:
        if line['Q7']==1 or line['Q8']==1 or line['Q9']==1:
            if line['Q10']==1 or line['Q11']==1 or line['Q12']==1:
                return 'Screen Immediately'
        if line['Q1']==1 or line['Q2']==1:
            if line['Q3']==1 or line['Q4']==1 or line['Q5']==1 or line['Q6']==1:
                return 'Screen after one Month'
        return 'No Risk'
    if line['Age']<36:
        if line['Q6']==1 or line['Q7']==1 or line['Q8']==1 or line['Q9']==1 or line['Q10']==1 :
            if line['Q11']==1 or line['Q12']==1 or line['Q13']==1 or line['Q14']==1:
                return 'Screen Immediately'
        if line['Q1']==1:
            if line['Q2']==1 or line['Q3']==1 or line['Q4']==1 or line['Q5']==1:
                return 'Screen after one Month'
        return 'No Risk'
    if line['Age']<48:
        if line['Q7']==1 or line['Q8']==1 or line['Q9']==1 or line['Q10']==1 or line['Q11']==1:
            if line['Q12']==1:
                return 'Screen Immediately'
        if line['Q1']==1:
            if line['Q2']==1 or line['Q3']==1 or line['Q4']==1 or line['Q5']==1 or line['Q6']==1:
                return 'Screen after one Month'
        return 'No Risk'
```

```python
result_list=[]
for i in range(len(df)):
    result_list.append(result_condition(df.iloc[i]))
    
```

```python
df['Result']=result_list
df.head()
```
